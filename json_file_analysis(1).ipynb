```python
import os
import re
import json
import gradio as gr
import torch
from langchain.schema import Document
from langchain.document_loaders import JSONLoader
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

# ==== Load model once globally ====
def load_model():
    model_id = "VibhorBhatt/gemma-funsd-finetuned"
    tokenizer = AutoTokenizer.from_pretrained(model_id, trust_remote_code=True)
    model = AutoModelForCausalLM.from_pretrained(
        model_id,
        device_map="cpu",  # set to "auto" if GPU available
        trust_remote_code=True
    )
    pipe = pipeline(
        "text-generation",
        model=model,
        tokenizer=tokenizer,
        max_new_tokens=128,
        temperature=0.1,
        top_p=0.9,
        do_sample=False,
        pad_token_id=tokenizer.eos_token_id,
        eos_token_id=tokenizer.eos_token_id
    )
    return pipe

llm_pipeline = load_model()

# ==== Extract Q&A pairs with bounding boxes ====
def extract_qa_from_form(json_data):
    id_map = {item["id"]: item for item in json_data.get("form", [])}
    qa_docs = []

    def get_combined_box(words):
        if not words:
            return []
        x0 = min(w["box"][0] for w in words if "box" in w)
        y0 = min(w["box"][1] for w in words if "box" in w)
        x1 = max(w["box"][2] for w in words if "box" in w)
        y1 = max(w["box"][3] for w in words if "box" in w)
        return [x0, y0, x1, y1]

    for item in json_data.get("form", []):
        if item.get("label") == "question" and item.get("linking"):
            for link in item["linking"]:
                question_id, answer_id = link
                question_item = id_map.get(question_id, {})
                answer_item = id_map.get(answer_id, {})

                question = question_item.get("text", "").strip()
                answer = answer_item.get("text", "").strip()

                # ✅ Use words[] to extract bounding box
                question_box = get_combined_box(question_item.get("words", []))
                answer_box = get_combined_box(answer_item.get("words", []))

                if question and answer:
                    qa_entry = {
                        "text": f"{question} {answer}",
                        "question_bbox": question_box,
                        "answer_bbox": answer_box
                    }
                    qa_docs.append(qa_entry)
    return qa_docs

# ==== Smart JSON schema detection ====
def smart_json_loader(file_path):
    candidate_schemas = [
        '.fullTextAnnotation.text',
        '.form[].text',
        '.data.text',
        '.text',
        '.pages[].text',
        '.document.text',
        '.results[].text'
    ]
    for schema in candidate_schemas:
        try:
            loader = JSONLoader(
                file_path=file_path,
                jq_schema=schema,
                text_content=False
            )
            data = loader.load()
            if data and any(d.page_content.strip() for d in data):
                print(f"✅ Loaded using schema: {schema}")
                return data, schema
        except Exception:
            continue
    raise ValueError("❌ No valid text found using known schemas.")

# ==== Main QA logic ====
def answer_question(file_obj, query):
    file_path = file_obj.name

    try:
        data, schema = smart_json_loader(file_path)
    except Exception as e:
        return f"❌ Failed to parse JSON: {str(e)}"

    bounding_box_info = ""

    if schema == '.form[].text':
        try:
            with open(file_path, "r") as f:
                raw_json = json.load(f)
            qa_pairs = extract_qa_from_form(raw_json)
            if not qa_pairs:
                return "❌ No Q&A pairs found in form[]."
            context = "\n".join([item['text'] for item in qa_pairs])

            # Also collect bounding box info for output
            for idx, pair in enumerate(qa_pairs):
                bounding_box_info += f"\nQ{idx+1} Bounding Boxes:\nQuestion Box: {pair['question_bbox']}\nAnswer Box: {pair['answer_bbox']}\n"

        except Exception as e:
            return f"❌ Error reading form data: {str(e)}"
    else:
        context = " ".join([doc.page_content.strip() for doc in data])

    if not context.strip():
        return "❌ No text found in document."

    # Add regex as hint
    matches = re.findall(r"\(?\d{3}\)?[-.\s]?\d{3}[-.\s]?\d{4}", context)
    phone_hint = f"📌 Found phone-like numbers: {', '.join(matches)}" if matches else ""

    # Prepare prompt
    prompt = f"""
You are an intelligent assistant. Based on the following document, answer the user's question as precisely as possible.

If there are multiple answers, return **only the most relevant one** based on clarity and closeness to the question.

If the answer cannot be found, respond with: "Not found".

---

Document:
{context}

Question: {query}
Answer (only the most relevant one):"""

    try:
        result = llm_pipeline(prompt)
        return result[0]["generated_text"] + (f"\n\n{phone_hint}" if phone_hint else "") + (f"\n\n📦 Bounding Box Info:{bounding_box_info}" if bounding_box_info else "")
    except Exception as e:
        return f"❌ Model error: {str(e)}"

# ==== Gradio UI ====
def create_interface():
    with gr.Blocks() as demo:
        gr.Markdown("## 🧠 Smart OCR QA App with Schema Detection + Bounding Boxes")

        with gr.Row():
            file_input = gr.File(label="Upload OCR JSON", type="filepath")

        with gr.Row():
            question_input = gr.Textbox(label="Ask a Question", placeholder="e.g. What is the fax number?")

        submit_btn = gr.Button("Get Answer")
        output = gr.Textbox(label="Answer with Bounding Box")

        submit_btn.click(fn=answer_question, inputs=[file_input, question_input], outputs=output)

    return demo

# ==== Launch ====
if __name__ == "__main__":
    demo = create_interface()
    demo.launch(share=True)
```

```python
#****
# 📌 First time only (in Colab):
# !pip install transformers gradio langchain pdf2image google-auth google-api-python-client poppler-utils
# adding api for analysis
import os
import re
import io
import json
import base64
import tempfile
import requests
import gradio as gr
from PIL import Image
from pdf2image import convert_from_path
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
from langchain.document_loaders import JSONLoader

# 🔐 Set API key (store in environment)
os.environ["AZURE_AKS_API_KEY"] = "AIzaSyDfMEbU2K4mwcagJXg_6ES6AyLvHNBAh3k"  # replace with actual key

# ✅ Google Vision API OCR wrapper
class getFromCnn:
    def __init__(self):
        api_key = os.environ.get('AZURE_AKS_API_KEY')
        if not api_key:
            raise ValueError("❌ API key not set.")
        self.url = f'https://vision.googleapis.com/v1/images:annotate?key={api_key}'

    def connect(self, image_bytes):
        if len(image_bytes) > 41943040:
            print("🚫 Image exceeds 40MB limit")
            return None

        json_data = {
            "requests": [{
                "features": [{"maxResults": 50, "model": "builtin/latest", "type": "DOCUMENT_TEXT_DETECTION"}],
                "image": {"content": base64.b64encode(image_bytes).decode('utf-8')},
                "imageContext": {"cropHintsParams": {"aspectRatios": [0.8, 1, 1.2]}}
            }]
        }

        try:
            print("📤 Sending OCR request...")
            response = requests.post(self.url, json=json_data)
            print("📥 Status:", response.status_code)
            print("🔎 Snippet:", response.text[:200])
            if response.status_code == 200:
                return response.json()
        except Exception as e:
            print(f"❌ Exception: {str(e)}")
        return None

# ✅ Convert PDF page to image bytes
class PageImageProcessor:
    def __init__(self, file_path):
        self.file_path = file_path

    def get_image_bytes(self, pagenum=0):
        print(f"🖼️ Converting page {pagenum} of PDF to image...")
        pages = convert_from_path(self.file_path)
        if pagenum >= len(pages):
            raise ValueError("Page number exceeds document.")
        buffer = io.BytesIO()
        pages[pagenum].save(buffer, format="PNG")
        return buffer.getvalue()

# ✅ Run Google Vision on image or PDF
def ocr_page_to_json(file_path, page=0):
    gcVision = getFromCnn()

    if file_path.lower().endswith((".png", ".jpg", ".jpeg")):
        # 🖼️ Image file
        with open(file_path, "rb") as img_file:
            image_bytes = img_file.read()
    elif file_path.lower().endswith(".pdf"):
        # 📄 Convert PDF page to image
        processor = PageImageProcessor(file_path)
        image_bytes = processor.get_image_bytes(page)
    else:
        raise ValueError("Unsupported file type. Please upload an image or PDF.")

    # 🔍 OCR via Google Vision
    resp = gcVision.connect(image_bytes)
    if not resp:
        raise Exception("❌ OCR failed: No response from API.")
    if "responses" not in resp or "fullTextAnnotation" not in resp["responses"][0]:
        raise Exception("❌ OCR failed or no text found.")

    # 📝 Save result to temporary JSON file
    text = resp["responses"][0]["fullTextAnnotation"]["text"]
    tmp_json = tempfile.NamedTemporaryFile(delete=False, suffix=".json")
    with open(tmp_json.name, "w") as f:
        json.dump({"fullTextAnnotation": {"text": text}}, f)

    print(f"✅ OCR result saved: {tmp_json.name}")
    return tmp_json.name


# ✅ Load fine-tuned Gemma
def load_model():
    model_id = "VibhorBhatt/gemma-funsd-finetuned"
    tokenizer = AutoTokenizer.from_pretrained(model_id, trust_remote_code=True)
    model = AutoModelForCausalLM.from_pretrained(
        model_id,
        device_map="auto",
        torch_dtype="auto",
        trust_remote_code=True,
        low_cpu_mem_usage=True
    )
    return pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=128)

llm_pipeline = load_model()

# ✅ JSON schema detection
def smart_json_loader(file_path):
    schemas = [
        '.fullTextAnnotation.text', '.form[].text', '.data.text', '.text',
        '.pages[].text', '.document.text', '.results[].text'
    ]
    for schema in schemas:
        try:
            loader = JSONLoader(file_path=file_path, jq_schema=schema, text_content=False)
            data = loader.load()
            if data and any(d.page_content.strip() for d in data):
                print(f"✅ Loaded schema: {schema}")
                return data
        except:
            continue
    raise ValueError("❌ No valid schema found.")

# ✅ Main handler
def process_document(file_obj, question, page_number=0):
    file_path = file_obj.name
    print(f"📂 Processing file: {file_path}")

    if file_path.endswith(('.pdf', '.png', '.jpg', '.jpeg')):
        try:
            page_number = int(page_number) if page_number else 0
            file_path = ocr_page_to_json(file_path, page_number)
        except Exception as e:
            return f"❌ OCR error: {str(e)}"

    try:
        data = smart_json_loader(file_path)
        merged_text = " ".join([doc.page_content.strip() for doc in data])
    except Exception as e:
        return f"❌ Text extraction failed: {str(e)}"

    if not merged_text:
        return "❌ Document contains no text."

    prompt = f"""You are an intelligent assistant. Read the document and answer the question. If not found, say "Not found".

Document:
{merged_text}

Question: {question}
Answer:"""

    result = llm_pipeline(prompt)
    return result[0]["generated_text"]

# ✅ Gradio Interface
with gr.Blocks() as demo:
    gr.Markdown("## 📄 OCR + Question Answering with Google Vision + Fine-tuned Gemma")
    file_input = gr.File(label="Upload PDF/Image or JSON")
    question_input = gr.Textbox(label="Ask a Question", placeholder="e.g. What is the FAX number?")
    page_number = gr.Number(label="Page # (for PDFs)", value=0, precision=0)
    submit = gr.Button("Get Answer")
    output = gr.Textbox(label="Answer")

    submit.click(fn=process_document, inputs=[file_input, question_input, page_number], outputs=output)

# ✅ Launch
demo.launch(share=True, debug=True)
```

```python
from datasets import load_dataset
dataset = load_dataset("nielsr/funsd")
```

```python
def extract_qa_from_form(sample):
    pairs = []
    # Check if 'form' key exists before accessing it
    if "form" in sample:
        id_map = {item["id"]: item for item in sample["form"]}
        for item in sample["form"]:
            if item["label"] == "question" and item["linking"]:
                for qid, aid in item["linking"]:
                    q = id_map.get(qid, {}).get("text", "").strip()
                    a = id_map.get(aid, {}).get("text", "").strip()
                    if q and a:
                        context = " ".join(w["text"] for w in item.get("words", []))
                        pairs.append(f"Document: {context}\\nQuestion: {q}\\nAnswer: {a}")
    return {"text": pairs}

qa_data = dataset["train"].map(extract_qa_from_form)
flattened = qa_data.flatten()
```

```python
from transformers import AutoTokenizer
model_id = "google/gemma-3-4b-it"  # or gemma-7b-it
tokenizer = AutoTokenizer.from_pretrained(model_id)

def tokenize_fn(example):
    # Filter out empty texts before tokenization
    #valid_texts = [text for text in example["text"] if text.strip()]  # Added strip() to remove whitespace
    valid_texts = [text for text in example["text"] if isinstance(text, str) and text.strip()] # Check if text is a string before stripping

    if not valid_texts:  # If all texts are empty, return empty dictionary
        return {}  # Return empty dict, which will be skipped by .map

    # Tokenize each valid text and pack into a single dictionary
    tokenized_texts = tokenizer(valid_texts, truncation=True, padding="max_length", return_tensors="pt")

    return tokenized_texts

tokenized = flattened.map(tokenize_fn, batched=True, remove_columns=['text'])
```

```python
from transformers import AutoModelForCausalLM, TrainingArguments, Trainer
from accelerate import Accelerator

# ✅ 1. Initialize Accelerate explicitly (prevents internal `.to()` calls)
accelerator = Accelerator()

# ✅ 2. Load model with offloading
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",            # Automatically splits layers
    offload_folder="offload",     # Disk offloading for layers
    torch_dtype="auto"            # Ensures compatibility with mixed precision
)

# ✅ 3. Gradient checkpointing helps with memory
model.gradient_checkpointing_enable()

# ✅ 4. Training arguments
args = TrainingArguments(
    output_dir="./gemma-funsd-finetuned",
    per_device_train_batch_size=2,
    num_train_epochs=3,
    logging_dir="./logs",
    save_strategy="epoch",
    fp16=True,
    remove_unused_columns=False,
    report_to="none"
)

# ✅ 5. Build trainer (DO NOT call `.to()` manually!)
trainer = Trainer(
    model=model,
    args=args,
    train_dataset=tokenized,
    tokenizer=tokenizer,
)

# ✅ 6. Start training (Accelerate handles device placement)
trainer.train()
```
